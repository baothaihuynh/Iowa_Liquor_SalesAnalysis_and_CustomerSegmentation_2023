`Comment:`  
- Looking at a small sample of data by summarizing it by 'Store Number', we can see that the 'Store Location' column has many unique data values due to differences from rounding decimal places in the store's location.  
- For this dataset, we will consider the geographical location as a type of data with relatively accurate precision, as there are many issues that can arise with this data, such as rounding errors, some places having the same longitude but different latitudes and vice versa, etc. Therefore, it is not possible to demand absolute accuracy in this information, but rather it will need to be retained with a certain level of relative accuracy, while also considering this during the data utilization process.  
- We will use the data from 'Store Location' to extract two new variables: 'Longitude' and 'Latitude', then remove the 'Store Location' column from the dataset.

```python
# Retrieve a dataframe named Location containing 'Store Number' and 'Store Location', then retain only the unique values based on 'Store Number'.
location = data[['Store Number', 'Store Location']].drop_duplicates(subset = 'Store Number')

# From 'Store Location', extract two new variables: 'Longitude' and 'Latitude', where the first element in each POINT of 'Store Location' corresponds to 'Longitude' and the remaining elements correspond to 'Latitude'.
location['Store Longitude'] = location['Store Location'].apply(lambda x: float(x.split()[1][1:]))
location['Store Latitude'] = location['Store Location'].apply(lambda x: float(x.split()[2][:-1]))

# Drop 'Store Location' column.
location = location.drop('Store Location', axis= 1)

print('Check samples location data:')
location.sample(5)
```

```python
print('Check unique data of location:')
location.shape
```

- The corresponding 'Longitude' and 'Latitude' positions for 2087 'Store Numbers' have been obtained.

```python
# Merge the two columns 'Longitude' and 'Latitude' with the 'data' dataframe using the common column 'Store Number'.
data = pd.merge(data, location, on = 'Store Number', how = 'left')

# Change the data type of 'Longitude' and 'Latitude'.
data['Store Longitude'] = data['Store Longitude'].astype('object')
data['Store Latitude'] = data['Store Latitude'].astype('object')

# Delete the unnecessary column 'Store Location'.
data = data.drop('Store Location', axis= 1)

data.head()
```

`Conclusion:`   
1. Analysis of Dispersion and Outliers:  
- Statistical measures:  
    - High standard deviation (std) and variance (var) indicate a wide dispersion of data.  
    - High skewness (skew) and kurtosis (kurtosis) indicate a non-symmetrical distribution and the presence of outliers.  
    - Large range (range) indicates a significant gap between the maximum (max) and minimum (min) values.  
- Visualizations:  
    - Distribution (distplot) and boxplot (boxplot) clearly show the large gap between the maximum and minimum data points.  
    - The majority of the data is concentrated at very low values, while a small number of outliers (around 4-6%) have significantly higher values than the rest.

2. Justification for Keeping Outliers:  
- Analysis objective: The project focuses on customer segmentation (Customer Segmentation). Therefore, outliers are important in explaining the characteristics of each customer group (e.g., stores with high order volumes, different alcohol consumption patterns across regions).

3. Handling Outliers in Descriptive Analysis:  
- Temporarily remove outliers to observe the distribution of the majority of the data.  
- The decision to remove or keep outliers will be carefully considered based on the analysis objective and the significance of each variable.  


```python
# Create new dataframe to drop outliers
data_no_outliers = data[num]

# List percentile of data no outliers for each feature
list_per = []
# Create a loop to drop outliers
for var in num:
    # Caculate Q1 & Q3
    Q1 = np.percentile(data_no_outliers[var], 25)
    Q3 = np.percentile(data_no_outliers[var], 75)

    # Find Lower Wisher & Upper Wisher
    LW = (Q1 - 1.5*scipy.stats.iqr(data_no_outliers[var]))
    UW = (Q3 + 1.5*scipy.stats.iqr(data_no_outliers[var]))

    # Get outliers value
    outliers = data_no_outliers[(data_no_outliers[var] < LW) | (data_no_outliers[var] > UW)][var]
    
    # Caculate percent of data when drop outliers for each features
    per = round((1 - len(outliers)/len(data_no_outliers[var]))*100,0)
    
    # Add that percent in list_per
    list_per.append(per)
    
    # Get min/max values (by features) of dataset when drop outliers
    data_no_outliers = data_no_outliers.drop(index= outliers.index)

# Check data distribution when drop outliers
i = 0
for col in num:
    print('Feature:', col)
    univariate_visualization_analysis_continuous_variable(data_no_outliers, data_no_outliers[col])
    univariate_analysis_continuous_variable(data_no_outliers, data_no_outliers[col])
    n_O_upper, n_L_upper, outliers_per = check_outlier(data_no_outliers, data_no_outliers[col])
    print("Percentage data of",col,"when drop outliers:", list_per[i],"%")
    print('\n')
    print('=' * 120 ,'\n')
    i += 1      
```
