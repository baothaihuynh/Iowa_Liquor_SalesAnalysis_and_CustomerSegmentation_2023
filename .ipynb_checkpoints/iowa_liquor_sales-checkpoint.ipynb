- Một số feature có kiểu dữ liệu không hợp lý cần thay đổi lại như: `Date`, `Store Number`, `Zip Code`, `Category`, `Vendor Number`, `Item Number`, `Pack`, `Bottles Sold`.  

```python
# Change date type

# Date
data['Date'] = pd.to_datetime(data['Date'])

# Store Number
data['Store Number'] = data['Store Number'].astype('int')
data['Store Number'] = data['Store Number'].astype('object')

# Zip Code
data['Zip Code'] = data['Zip Code'].astype('int')
data['Zip Code'] = data['Zip Code'].astype('object')

# Category
data['Category'] = data['Category'].astype('int')
data['Category'] = data['Category'].astype('object')

# Vendor Numbe
data['Vendor Number'] = data['Vendor Number'].astype('int')
data['Vendor Number'] = data['Vendor Number'].astype('object')

# Item Number
data['Item Number'] = data['Item Number'].astype('int')
data['Item Number'] = data['Item Number'].astype('object')

# Pack
data['Pack'] = data['Pack'].astype('int')
data['Pack'] = data['Pack'].astype('object')

# Bottles Sold
data['Bottles Sold'] = data['Bottles Sold'].astype('int')
```

```python
data.dtypes
```

```python
# Kiểm tra dữ liệu ở các biến Bottle Volumne (ml), State Bottle Cost, State Bottle Retail, Bottles Sold, Sale (Dollars), Volume Sold (Liters), Volumne Sold (Gallons) có tồn tại dữ liệu âm và bằng 0 hay không.
for col in data.select_dtypes(include= ['int', 'float']).columns:
    print('Kiểm tra biến:', col)
    count_negative_or_zero = data[data[col] <= 0][col].count()
    
    if count_negative_or_zero > 0:
        print('Số lượng dữ liệu bé hơn hoặc bằng 0:', count_negative_or_zero)
        print('Mẫu dữ liệu:')
        print(data[data[col] <= 0][data.select_dtypes(include= ['int', 'float']).columns].sample(3))
    else:
        print('Không có dữ liệu bé hơn hoặc bằng 0')
    
    print('\n')
    print('*' * 100)
    print('\n')
```

`Comment:`  
- Qua thông kê ở trên ta thấy được lượng dữ liệu bé hơn hoặc bằng 0 này nằm ở 4 biến Bottles Sold, Sale (Dollars), Volume Sold (Liters) và Volume Sold (Gallons).
- Đây có thể được xem như một lỗi sai trong quá trình cập nhật dữ liệu số dương ở biến Bottles Sold thành số âm, từ đó ảnh hưởng đến 3 biến còn lại ở phía sau do các phép tính toán giữa các biến.
- Hướng giải quyết đối với các dữ liệu phi logic ở trên là loại bỏ hoàn toàn những dòng dữ liệu bởi vì lượng dữ liệu sai chỉ chiếm một phần không đáng kể so với tổng thể.

```python
# Xóa dữ liệu bé hơn hoặc bằng 0
# Chọn cột 'Volume Sold (Gallons)' để lấy ra các dữ liệu bé hơn hoặc bằng 0 vì lượng dữ liệu thỏa điều kiện ở cột này chiếm số lượng nhiều nhất trong 4 cột
filter_data = data[data['Volume Sold (Gallons)'] <= 0]

# xóa dữ liệu bé hơn hoặc bằng 0
data = data.drop(filter_data.index)
data.reset_index(drop= True, inplace= True)
```

```python
# Kiểm tra dữ liệu sau khi thực hiện xóa dữ liệu bé hơn hoặc bằng 0
for col in data.select_dtypes(include= ['int', 'float']).columns:
    print('Kiểm tra biến:', col)
    count_negative_or_zero = data[data[col] <= 0][col].count()
    
    if count_negative_or_zero > 0:
        print('Số lượng dữ liệu bé hơn hoặc bằng 0:', count_negative_or_zero)
        print('Mẫu dữ liệu:')
        print(data[data[col] <= 0][data.select_dtypes(include= ['int', 'float']).columns].sample(3))
    else:
        print('Không có dữ liệu bé hơn hoặc bằng 0')
    
    print('\n')
    print('*' * 100)
    print('\n')
```

- Nhìn vào một mẩu nhỏ dữ liệu qua việc thống kê theo 'Store Number', có thể thấy cột dữ liệu 'Store Location' có nhiều dữ liệu duy nhất là do sự chênh lệch từ việc làm tròn phần số thập phân trong vị trí của cửa hàng.
- Đối với tập dữ liệu này, ta sẽ xem vị trí địa lí như một loại dữ liệu với độ chính xác ở mức tương đối, vì có rất nhiều vấn đề phát sinh ở dữ liệu này có thể kể đến như: sai số do làm tròn hàng thập phân, một số nơi có cùng kinh độ nhưng khác vĩ độ và ngược lại,... Từ đó không thể yêu cầu độ chính xác tuyệt đối ở thông tin này mà sẽ cần nhắc giữ lại với một mức độ tương đối, đồng thời cân nhắc trong quá trình sử dụng thông tin.
- Ta sẽ sử dụng dữ liệu từ 'Store Location' để lấy ra 2 biến mới là 'Longitude' và 'Latitude' sau đó loại bỏ cột 'Store Location' ra khỏi tập dữ liệu.

```python
# Lấy ra dataframe Location bao gồm 'Store Number' và 'Store Location' sau đó chỉ giữ lại các giá trị duy nhất theo 'Store Number'.
location = data[['Store Number', 'Store Location']].drop_duplicates(subset = 'Store Number')

# Từ 'Store Location', lấy ra 2 biến dữ liệu mới là 'Longitude' và 'Latitude'. Trong đó phần tử đầu tiên trong mỗi POINT của 'Store Location' tưởng ứng với 'Longitude' và phần tử còn lại tương ứng với 'Latitude'.
location['Store Longitude'] = location['Store Location'].apply(lambda x: float(x.split()[1][1:]))
location['Store Latitude'] = location['Store Location'].apply(lambda x: float(x.split()[2][:-1]))

# Xóa cột 'Store Location'.
location = location.drop('Store Location', axis= 1)

print('Check samples location data:')
location.sample(5)
```

```python
print('Check unique data of location:')
location.shape
```

- Đã có được vị trí 'Longitude' và 'Latitude' tương ứng của 2086 'Store Number'.  
- Tiếp theo sẽ tiến hành ghép 2 dữ liệu trên vào dataframe data.

```python
# Ghép 2 cột 'Longitude' và 'Latitude' và dataframe data bằng cột dữ liệu chung là 'Store Number'.
data = pd.merge(data, location, on = 'Store Number', how = 'left')

# Thay đổi kiểu dữ liệu của 'Longitude' và 'Latitude'
data['Store Longitude'] = data['Store Longitude'].astype('object')
data['Store Latitude'] = data['Store Latitude'].astype('object')

# Xóa cột 'Store Location' không cần thiết.
data = data.drop('Store Location', axis= 1)

data.head()
```
