## **III. Data Preparation**

#### 1. Select Features To Build Model

```python
products = data.groupby('Product Name').agg({'State Bottle Retail' : 'mean',
                                            'Bottles Sold' : 'sum',
                                            'Sale (Dollars)' : 'sum'}).reset_index()
products.sort_values(by = 'Sale (Dollars)', ascending = False, inplace = True)
products
```

#### 2. Summary About Dataframe Products

```python
for col in products.columns[1:]:
    print('Feature:', col)
    univariate_visualization_analysis_continuous_variable(products, products[col])
    univariate_analysis_continuous_variable(products, products[col])
    n_O_upper, n_L_upper, outliers_per = check_outlier(products, products[col])
    print('\n')
    print('*' * 60 ,'\n')
```

#### 3. Feature Engineering

```python
def analyze_skewness(x, df):
    fig, ax = plt.subplots(1,5, figsize=(10,5))
    sb.distplot(df[x], ax=ax[0], color= "#005A74")
    sb.distplot(np.log(df[x]), ax=ax[1], color= "#005A74")
    sb.distplot(np.sqrt(df[x]), ax=ax[2], color= "#005A74")
    sb.distplot(stats.boxcox(df[x])[0], ax=ax[3], color= "#005A74")
    sb.distplot(np.cbrt(df[x]), ax=ax[4], color= "#005A74") 
    plt.tight_layout()
    plt.show()
    #Print result 
    print("Original Skewness:", df[x].skew().round(2))
    print("Log-transformed Skewness:", np.log(df[x]).skew().round(2))
    print("Square Root-transformed Skewness:", np.sqrt(df[x]).skew().round(2))
    print("Box-Cox transformed Skewness:", pd.Series(stats.boxcox(df[x])[0]).skew().round(2))
    print("Cube Root-transformed Skewness:", pd.Series(np.cbrt(df[x])).skew().round(2))
```

```python
for col in products.columns[1:]:
    analyze_skewness(col, products)
```

#### 4. Create Products Dataframe with Boxcox

```python
products_bc = pd.DataFrame()
# Tranform Recency to box-cox transformation
products_bc['State Bottle Retail'] = stats.boxcox(products['State Bottle Retail'])[0]

# Tranform Frequency to box-cox transformation
products_bc['Bottles Sold'] = stats.boxcox(products['Bottles Sold'])[0]

# Tranform Monetary to box-cox transformation
products_bc['Sale (Dollars)'] = stats.boxcox(products['Sale (Dollars)'])[0]

# Check new dataframe
products_bc.head()
```

#### 5. Scaler Data with Standard Scaler

```python
# Using StandardScaler
scaler = SklearnTransformerWrapper(transformer = StandardScaler())
scaler.fit(products_bc)
products_scaler = scaler.transform(products_bc)
products_scaler
```

```python
for col in products.columns[1:]:
    print('Feature:', col)
    univariate_visualization_analysis_continuous_variable(products_scaler, products_scaler[col])
    univariate_analysis_continuous_variable(products_scaler, products_scaler[col])
    n_O_upper, n_L_upper, outliers_per = check_outlier(products_scaler, products_scaler[col])
    print('\n')
    print('*' * 60 ,'\n')
```

## **IV. Products Segmentation with Machine Learning**

#### 1. Using K-Means with Elbow Method

```python
# Loop from 2 to 10 clusters and monitor the WSSE (within cluster sum of squares)
wsse_list = []
K_range = range(2, 11)

for k in K_range:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k, random_state=42)

    # Fit the model to the data
    model.fit(products_scaler)

    # Append the inertia to the list of inertias
    wsse_list.append(model.inertia_)
```

```python
# Visualize the elbow plot
plt.figure(figsize=(10, 6))
sb.lineplot(x=K_range, y=wsse_list, marker='o', color= "#005A74")
plt.xlabel('Number of clusters')
plt.ylabel('WSSE')
plt.title('Elbow Plot',loc = 'left',  fontweight = 'heavy', fontsize = 16)
plt.show();
```

#### 2. Using Hierarchical Clustering with Dendrogram

```python
# Calculate the linkage
linkage_matrix = linkage(products_scaler, method = 'ward', metric= 'euclidean')
```

```python
# Plot the dendrogram
plt.figure(figsize = (15, 10))
dendrogram(linkage_matrix, leaf_rotation= 90, leaf_font_size= 6)
plt.title('Dendrogram',loc = 'left',  fontweight = 'heavy', fontsize = 16)
plt.xlabel('News')
plt.ylabel('Distance')
plt.show();
```

#### 3. Using K-Means with Silhouette Method

```python
silhouette_scores = []
possible_k_values = range(2, 11)  # k from 2 to 10
for k in possible_k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(products_scaler)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(products_scaler, labels))

# Visualize Silhouette Score for each k value
plt.figure(figsize=(10, 6))
plt.plot(possible_k_values, silhouette_scores, marker='o', color= "#005A74")
plt.title('Silhouette Score for Different k',loc = 'left',  fontweight = 'heavy', fontsize = 16)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(possible_k_values)
plt.grid(True)
plt.show()

# Choose the best k
best_k = possible_k_values[np.argmax(silhouette_scores)]
print(f"Best k value based on Silhouette Score: {best_k}")
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
# Visualize
fig, ax = plt.subplots(2, 2, figsize=(15, 9))
for idx, k in enumerate([2, 3, 4, 5]):
    km = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(idx, 2)

    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q][mod])
    visualizer.fit(products_scaler)
    ax[q][mod].text(0.5, -0.1, f'Silhouette Score: {np.mean(visualizer.silhouette_score_):.2f}', size=12, ha='center', transform=ax[q][mod].transAxes)
plt.tight_layout()
plt.show()
```

```python
k = 4
kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=1000, random_state=42)
kmeans = kmeans.fit(products_scaler)
clusters = kmeans.predict(products_scaler)
products_scaler['Cluster'] = clusters

# Visualize with 3D graph by Recency, Frequency and Monetary
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')

for i in range(k):
    cluster_data = products_scaler[products_scaler['Cluster'] == i]
    ax.scatter(cluster_data['State Bottle Retail'], cluster_data['Bottles Sold'], cluster_data['Sale (Dollars)'], label=f'Cluster {i}')

ax.set_xlabel('State Bottle Retail (Scaled)')
ax.set_ylabel('Bottles Sold (Scaled)')
ax.set_zlabel('Sale (Dollars) (Scaled)')
ax.set_title('RFM Clustering (3D Scatter Plot)',loc = 'left',  fontweight = 'heavy', fontsize = 16)
ax.legend()
plt.show()
```
