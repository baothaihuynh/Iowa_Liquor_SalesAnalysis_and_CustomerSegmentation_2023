```python
def univariate_analysis_continuous_variable(df, feature):
    print("Describe:")
    print(feature.describe(include = 'all'))
    print("Mode:", feature.mode())
    print("Range:", feature.values.ptp())
    print("IQR:", scipy.stats.iqr(feature))
    print("Var:", feature.var())
    print("Skew:", feature.skew())
    print("Kurtosis:", feature.kurtosis())
```

```python
def univariate_visualization_analysis_continuous_variable(df,feature):
    sb.set()
    fig, (ax0, ax1) = plt.subplots(1,2, figsize = (15,5) )  
    sb.distplot(feature, ax= ax0)
    sb.boxplot(data= df, x = feature, ax = ax1)
    plt.show()
```

```python
#Number of upper, lower outliers
def check_outlier(df, feature):
    # sb.boxplot(data = df, y = feature)
    # plt.show()
    Q1 = np.percentile(feature, 25)
    Q3 = np.percentile(feature, 75)
    n_O_upper = df[feature > (Q3 + 1.5*scipy.stats.iqr(feature))].shape[0]
    print("Number of upper outliers:", n_O_upper)
    n_L_upper = df[feature < (Q1 - 1.5*scipy.stats.iqr(feature))].shape[0]
    print("Number of lower outliers:", n_L_upper)
    #Percentage of outliers
    outliers_per = round(((n_O_upper + n_L_upper)/df.shape[0])*100,0)
    print("Percentage of outliers:", outliers_per,"%")
    return  n_O_upper, n_L_upper, outliers_per
```

```python
# Numerical fetures
num = ['State Bottle Cost', 'State Bottle Retail', 'Bottles Sold', 'Sale (Dollars)', 'Volume Sold (Liters)']
```

```python
for col in num:
    print('Feature:', col)
    univariate_visualization_analysis_continuous_variable(data, data[col])
    univariate_analysis_continuous_variable(data, data[col])
    n_O_upper, n_L_upper, outliers_per = check_outlier(data, data[col])
    print('\n')
    print('*' * 60 ,'\n')
```

`Comment:`  
- Thông qua việc sử dụng các chỉ số thống kê kết hợp với trực quan hóa dữ liệu của từng biến số liên tục, ta nhận thấy một số vấn đề ở các biến như sau:  
    - Các chỉ số thống kê như `std`, `var`, `skew`, `kurtosis` tương đối cao cho thấy độ phân tán của dữ liệu khá rộng. Nhìn vào `range` cho thấy khoảng cách giữa `max` và `min` là rất lớn.  
    - Biểu đồ `distplot` và `boxplot` thể hiện rõ hơn về khoảng cách giữa dữ liệu lớn nhất và dữ liệu nhỏ nhất và điều này xuất hiện ở tất cả các biến. Phần lớn dữ liệu tập trung ở các giá trị rất nhỏ, trong khi đó  một số ít dữ liệu outliers khoảng 4-6% đạt giá trị vượt bậc so với phần còn lại.    
- Tuy nhiên, ta sẽ không loại bỏ outliers trong bài phân tích bởi vì project này đang hướng đến giải quyết bài toán `Customer Segmentation` nên các outliers trên đang mang nhiều ý nghĩa giải thích đặc trưng của từng nhóm khách hàng (ví dụ: Một số cửa hàng có lượng đặt hàng cao hơn, Nhu cầu tiêu thụ các loại rượu khác nhau qua từng khu vực,…).  
- Ở quá trình phân tích mô tả của các biến số, ta sẽ chỉ loại bỏ tạm thời các outliers để xem phần lớn dữ liệu đang được phân bố như thế nào.

```python
# Tạo dataframe dùng để loại bỏ các outliers
data_no_outliers = data[num]

# List percentile of data no outliers for each feature
list_per = []
# Tạo vòng lặp loại bỏ outliers
for var in num:
    #Tính Q1 và Q3
    Q1 = np.percentile(data_no_outliers[var], 25)
    Q3 = np.percentile(data_no_outliers[var], 75)

    #Tìm Lower Wisher và Upper Wisher
    LW = (Q1 - 1.5*scipy.stats.iqr(data_no_outliers[var]))
    UW = (Q3 + 1.5*scipy.stats.iqr(data_no_outliers[var]))

    #Lấy ra các giá trị outlier
    outliers = data_no_outliers[(data_no_outliers[var] < LW) | (data_no_outliers[var] > UW)][var]
    
    # Tính tỷ lệ dữ liệu còn lại sau khi xóa outliers của mỗi biến
    per = round((1 - len(outliers)/len(data_no_outliers[var]))*100,0)
    
    # Thêm tỷ lệ vừa tính được ở trên vào danh sách list_per
    list_per.append(per)
    
    #Lấy ra các giá trị min/max (theo từng biến) của tập dữ liệu sau khi loại bỏ outlier
    data_no_outliers = data_no_outliers.drop(index= outliers.index)

# Xem phân bố dữ liệu sau khi xóa outlier
i = 0
for col in num:
    print('Feature:', col)
    univariate_visualization_analysis_continuous_variable(data_no_outliers, data_no_outliers[col])
    univariate_analysis_continuous_variable(data_no_outliers, data_no_outliers[col])
    n_O_upper, n_L_upper, outliers_per = check_outlier(data_no_outliers, data_no_outliers[col])
    print("Percentage data of",col,"when drop outliers:", list_per[i],"%")
    print('\n')
    print('=' * 120 ,'\n')
    i += 1      
```

```python
# Tỷ lệ dữ liệu còn lại sau khi loại bỏ outliers
print('Percentage of total dataset when drop outliers:',round((data_no_outliers.shape[0]/data.shape[0])*100,0),'%')
```

```python
# Group 'Sale (Dollars)' of each 'Store Name' by 'Month'.
cus_purchase_month = data.groupby(['Store Name', 'Year Month'])['Sale (Dollars)'].sum().reset_index()
cus_purchase_month = cus_purchase_month.rename(columns = {'Sale (Dollars)' : 'Revenue ($)'})
cus_purchase_month.sort_values(by = 'Store Name')
```

```python
# Tạo dataframe với index là 'Store Name' tương ứng với 12 tháng
# Tháng mang giá trị 1 là tháng có mua hàng, 0 là không mua hàng
df_retention = pd.crosstab(cus_purchase_month['Store Name'], cus_purchase_month['Year Month'])
df_retention = df_retention.rename_axis(None, axis = 1).reset_index()
df_retention.columns = df_retention.columns.astype('str')
df_retention
```

```python
# Lấy ra danh sách tháng
months = df_retention.columns[1:]
months
```

```python
#Tạo một list retention, trong đó mỗi phần tử là một dictionary gồm 3 Key: 'Year MOnth', 'Total Customers' và 'Retained Customers'.
# Ta sẽ không thực hiện phép tính cho tháng 1 do tháng 1 không có dữ liệu tháng liền kề phía trước.
retention_array = []

# Cho chạy vòng lặp 11 lần vì tháng 1 sẽ không có tháng liền kề trước đó nên vòng lặp sẽ chạy từ tháng 2 cho tới 12
for i in range(len(months) - 1):
    # Tạo một dictionary rỗng để mỗi vòng lặp sẽ gán vào dữ liệu mới
    retention_data = {}

    # Lấy ra tháng hiện tại, bắt đầu từ tháng 2 và tăng dần sau mỗi lần lặp
    selected_month = months[i+1]

    # Lấy ra tháng phía trước liền kề
    prev_month = months[i]

    # Tạo key là 'Invoice Year Month' với value tương ứng là tháng hiện tại
    retention_data['Year Month'] = selected_month

    # Tạo key là 'Total Customers Count' với value tương ứng là tổng số khách hàng mua hàng trong tháng hiện tại
    retention_data['Total Customers'] = df_retention[selected_month].sum()

    # Tạo key là 'Retained Customers Count' với value tương ứng là tổng số khách hàng vừa mua hàng trong tháng hiện tại đồng thời cũng mua hàng ở tháng liền kề trước đó
    retention_data['Retained Customers'] = df_retention[(df_retention[selected_month] == 1) & (df_retention[prev_month] == 1)][selected_month].sum()

    # Thêm dictionary vào list retention đã tạo ban đầu
    retention_array.append(retention_data)
    print("**********************************************************" + selected_month + "**********************************************************")
    print(retention_array, '\n')
```

```python
# Sau khi kết thúc vòng lặp ở trên, ta có được một array bao gồm các dictionary
# Từ đó ta tạo nên dataframe retention bằng dữ liệu trong retention_array
retention = pd.DataFrame(retention_array)

# Thực hiện tính Retention Rate như sau:
#         Customers Retention Rate = (Retained Customers / Total Customers) * 100
# Trong đó:
#         Retained Customers là các khách hàng đã mua hàng của tháng liền kề trước và tiếp tục mua hàng trong tháng hiện tại.
#         Total Customers là tổng số khách hàng của tháng hiện tại (bao gồm khách hàng của tháng trước tiếp tục mua vào tháng hiện tại và khách hàng mới mua trong tháng).
retention['Customers Retention Rate (%)'] = round((retention['Retained Customers'] / retention['Total Customers']) * 100, 2)
retention
```
